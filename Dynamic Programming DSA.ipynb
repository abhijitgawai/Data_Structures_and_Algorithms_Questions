```python
'''GFG'''
def knapSack(W, wt, val, n):
	dp = [0 for i in range(W+1)] # Making the dp array

	for i in range(1, n+1): # taking first i elements
		for w in range(W, 0, -1): # starting from back,so that we also have data of
								# previous computation when taking i-1 items
			if wt[i-1] <= w:
				# finding the maximum value
				dp[w] = max(dp[w], dp[w-wt[i-1]]+val[i-1])
	return dp[W] 

val = [60, 100, 120]
wt = [10, 20, 30]
W = 50
n = len(val)
print(knapSack(W, wt, val, n))                                               # learn this way to solve in 0(n) space and time. others is O(n^2)
```

```python
'''GFG O(N) ''' 
int cutRod(int price[], int n) {
    vector<int> dp(n + 1, 0);
    dp[0] = 0;
    for (int i = 1; i <= n; i++) {
        for (int j = 0; j < i; j++) {
            dp[i] = max(dp[i], dp[j] + price[i - j - 1]);
        }
    }
    return dp[n];
}
```

```python
def minimumDeleteSum(self, s1: str, s2: str) -> int:

    t=[[0 for i in range(len(s2)+1)] for j in range(len(s1)+1)]

    for i in range(1,len(t)):
        for j in range(1,len(t[0])):
            if s2[j-1]==s1[i-1]:
                t[i][j]=ord(s1[i-1])+t[i-1][j-1]
            else:
                t[i][j]=max(t[i-1][j],t[i][j-1])


    re=0
    for i in range(len(s1)):
        re+=ord(s1[i])
    for i in range(len(s2)):
        re+=ord(s2[i])

    return re-(2*(t[-1][-1]))
```

```python
#fastest
class Solution:
    # @param A : string
    # @return an integer
    def minCut(self, A):
        le=len(A)
        dp=[i for i in range(le)]
        for i in range(le):
            l=i
            r=i
            while(l>=0 and r<le):
                if(A[l]==A[r]):
                    if(l!=0):
                        dp[r]=min(dp[l-1]+1,dp[r])
                    else:
                        dp[r]=min(dp[r],0)
                else:
                    break
                l-=1
                r+=1
            l=i-1
            r=i
            while(l>=0 and r<le):
                if(A[l]==A[r]):
                    if(l!=0):
                        dp[r]=min(dp[l-1]+1,dp[r])
                    else:
                        dp[r]=min(dp[r],0)
                else:
                    break
                l-=1
                r+=1
        return dp[-1]
```

```python
#light weight
class Solution:
    # @param A : string
    # @return an integer
    def minCut(self, A):
        def isPalin(string):
            for i in range(0, int(len(string)/2)):
                if string[i] != string[len(string)-1-i]:
                    return False
            return True

        if len(A) == 0 or len(A) == 1:
            return 0
        
        min_cuts = [len(A)+1] * (len(A)+1)
        min_cuts[0] = 0
        
        for i in range(1, len(A)+1):
            if isPalin(A[0:i]):
                min_cuts[i] = 0
            else:
                for j in range(0,i):
                    if isPalin(A[j:i]) and min_cuts[j] + 1 < min_cuts[i]:
                        min_cuts[i] = min_cuts[j] + 1
        
        return min_cuts[len(A)]
```

# 169	Diameter of a tree


```python
class Solution:
    
    def solve(self,root):
        if not root:
            return 0
        
        l=self.solve(root.left)
        r=self.solve(root.right)
        
        temp=max(l,r)+1
        ans=max(temp,l+r+1)
        self.res=max(ans,self.res)
        return temp
        
    def diameter(self,root):

        if not root:
            return 0
        self.res=0
        self.solve(root)
        return self.res
    
'''This is also correct, but TLE'''    
# def Height(self,r): 
#     if r is None:
#         return 0
#     lef=self.Height(r.left)
#     rig=self.Height(r.right)
#     return max(lef,rig)+1


# def diameter(self,root):
#     if root == None:
#             return 0
#     r=root
#     rig=self.Height(r.right)
#     lef=self.Height(r.left)
#     op1=rig+lef+1
#     op2=self.diameter(r.right)
#     op3=self.diameter(r.left)
#     return max(op1,op2,op3)
```

# Binary Tree Maximum Path Sum (Leetcode 124)


```python
import sys

class Solution:
    def solve(self,root):
        
        if root==None:
            return 0
        
        l=self.solve(root.left)
        r=self.solve(root.right)
        
        temp=max(max(l,r)+root.val,root.val)
        ans=max(temp,l+r+root.val)
        self.res=max(self.res,ans)
        
        return temp
        
        
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        self.res=-sys.maxsize+1
        if root==None:
            return 0
        #self.res=0
        self.solve(root)
        
        return max(self.res,root.val)
```
